```
#TRYING MONTH
montha = dfDropDup[['Visitor_ID','Trying_Months']].copy()
#month = montha.fillna(0)
monthNoDup = montha.drop_duplicates()
monthDetectChange = pd.DataFrame(monthNoDup.groupby('Visitor_ID').agg({'Trying_Months' : Changes}).reset_index().values, columns=["Visitor_ID", "Month Changes"])
dfMonthMerge = pd.merge(montha[['Visitor_ID','Trying_Months']],monthDetectChange[['Visitor_ID','Month Changes']], on = 'Visitor_ID', how = 'left')

#if changes = 0 & 1: Keep
dfMonthMerge_0 = pd.DataFrame(dfMonthMerge.loc[dfMonthMerge['Month Changes'] < 2].reset_index().values, columns=["Unnamed: 0","Visitor_ID","Trying Month New", "Month Changes"])
#if changes >=2: remove the default value of 0
dfMonthMerge_2 = dfMonthMerge.loc[dfMonthMerge['Month Changes'] > 1]
dfMonthMerge_2 = dfMonthMerge_2.fillna(0)
month_Remove = (dfMonthMerge_2['Trying_Months'] != 0).groupby(dfMonthMerge_2['Visitor_ID']).cumsum()
month_Remove2 = pd.DataFrame(dfMonthMerge_2[month_Remove!=0].reset_index().values, columns=["Unnamed: 0","Visitor_ID","Trying Month New", "Month Changes"])

#merge all changes
month_final = dfMonthMerge_0.append(month_Remove2)

dfRemoveMonth = pd.merge(dfDropDup,month_final[['Unnamed: 0','Trying Month New']], on = 'Unnamed: 0', how='left')
dfRemoveMonth

#from google.colab import drive
#drive.mount('drive')
#df831a.to_csv('df831a.csv')
#!cp df831a.csv "/content/drive/MyDrive/05. STUDY/FLINDERS/2020-08 Masters Thesis/"
```

```
#SMOKING
Smokinga = dfDropDup[['Visitor_ID','Smoking']].copy()
Smoking = Smokinga.fillna(0) #assume no NaN for this attribute
SmokingNoDup = Smoking.drop_duplicates()
SmokingDetectChange = pd.DataFrame(SmokingNoDup.groupby('Visitor_ID').agg({'Smoking' : Changes}).reset_index().values, columns=["Visitor_ID", "Smoking Changes"])
dfSmokingMerge = pd.merge(Smoking[['Visitor_ID','Smoking']],SmokingDetectChange[['Visitor_ID','Smoking Changes']], on = 'Visitor_ID', how = 'left')

#if changes = 0: Keep
dfSmokingMerge_0 = pd.DataFrame(dfSmokingMerge.loc[dfSmokingMerge['Smoking Changes'] < 1].reset_index().values, columns=["Unnamed: 0","Visitor_ID","Smoking New", "Smoking Changes"])
#if changes >0: remove the default value of 0
dfSmokingMerge_2 = dfSmokingMerge.loc[dfSmokingMerge['Smoking Changes'] > 0]
Smoking_Remove = (dfSmokingMerge_2['Smoking'] != 0).groupby(dfSmokingMerge_2['Visitor_ID']).cumsum()
Smoking_Remove2 = pd.DataFrame(dfSmokingMerge_2[Smoking_Remove!=0].reset_index().values, columns=["Unnamed: 0","Visitor_ID","Smoking New", "Smoking Changes"])

#merge all changes
Smoking_final = dfSmokingMerge_0.append(Smoking_Remove2)

dfRemoveSmoking = pd.merge(dfRemoveMonth,Smoking_final[['Unnamed: 0','Smoking New']], on = 'Unnamed: 0', how='left')
dfRemoveSmoking
```
